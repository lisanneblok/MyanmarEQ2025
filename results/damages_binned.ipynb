```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
```

# Pair 20

```python
# Path to the shapefile
shapefile_path = 'data/pair20_damaged_output_polygons.shp'

# Read the shapefile into a GeoDataFrame
gdf_building_20 = gpd.read_file(shapefile_path)

# Display the first few rows of the GeoDataFrame to check
print(gdf_building_20.plot())
```

```python
# Replace all NaN values in the entire GeoDataFrame with 0
gdf_building_20.mean_value.fillna(0, inplace=True)
# Replace all NaN values in the entire GeoDataFrame with 0
gdf_building_20.normalized.fillna(0, inplace=True)
```

```python
# Create bins of size 0.1 from 0 to 1 and round the edges to the nearest decimal
bins = [round(i * 0.1, 1) for i in range(11)]  # This will create bins [0.0, 0.1, 0.2, ..., 1.0]
labels = [f'{round(i*0.1, 1)}-{round((i+1)*0.1, 1)}' for i in range(10)]  # Labels for the bins

# Use pd.cut to bin the normalized_mean_value into specified bins
gdf_building_20['binned'] = pd.cut(gdf_building_20['normalized'], bins=bins, labels=labels, include_lowest=True)

# Count the number of buildings in each bin
bin_counts = gdf_building_20['binned'].value_counts().sort_index()

# Create a summary DataFrame with the counts for each bin
summary_df_20 = pd.DataFrame(bin_counts).reset_index()

# Apply pd.cut to bin the 'normalized' column values into the specified bins
gdf_building_20['bin_label'] = pd.cut(gdf_building_20['normalized'], bins=bins, labels=labels, include_lowest=True)

print(summary_df_20)
```

```python
# Filter the GeoDataFrame to get only rows where 'binned' is '0.9-1.0'
filtered_gdf = gdf_building_20[gdf_building_20['binned'] == '0.5-0.6']

# Get the centroid of the first 10 geometries and print them
centroids = filtered_gdf.geometry.centroid
# Iterate and print each centroid
for centroid in centroids:
    print(centroid)
```

# Pair 2

```python
# Path to the shapefile
shapefile_path = 'data/pair2_damaged_output_polygons.shp'

# Read the shapefile into a GeoDataFrame
gdf_building_2 = gpd.read_file(shapefile_path)
```

```python
# Create bins of size 0.1 from 0 to 1 and round the edges to the nearest decimal
bins = [round(i * 0.1, 1) for i in range(11)]  # This will create bins [0.0, 0.1, 0.2, ..., 1.0]
labels = [f'{round(i*0.1, 1)}-{round((i+1)*0.1, 1)}' for i in range(10)]  # Labels for the bins

# Use pd.cut to bin the normalized_mean_value into specified bins
gdf_building_2['binned'] = pd.cut(gdf_building_2['normalized'], bins=bins, labels=labels, include_lowest=True)

# Count the number of buildings in each bin
bin_counts = gdf_building_2['binned'].value_counts().sort_index()

# Create a summary DataFrame with the counts for each bin
summary_df_2 = pd.DataFrame(bin_counts).reset_index()

# Apply pd.cut to bin the 'normalized' column values into the specified bins and set bin_label to the lower bound of each bin
gdf_building_2['bin_label'] = pd.cut(gdf_building_2['normalized'], bins=bins, labels=False, include_lowest=True).apply(lambda x: bins[x])

print(summary_df_2)
```
